```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris =pd.read_csv('praktikum/covid_19_indonesia.csv')
```

```python
iris.head()
```

```python
df=iris.drop(['Date','Location ISO Code'],axis=1)
df.head()
```

```python
plt.plot(df['New Cases'])
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Data')
plt.show()
```

```python
plt.plot(df['New Cases'],color='yellow',marker='+')
plt.xlabel('Number')
plt.ylabel('Sepal Lengh')
plt.title('Data')
plt.show()
```

```python
plt.plot(df['New Cases'],'--c')
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
plt.show()
```

```python
plt.plot(df['New Cases'],'--c')
plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Iris Data')
limits=[0,50,0,10]
plt.show()
```

```python
plt.plot(df['New Cases'])
plt.plot(df['Total Cases'])

plt.plot(df['New Deaths'])
plt.plot(df['Total Deaths'])

plt.plot(df['New Recovered'])
plt.plot(df['Total Recovered'])

plt.plot(df['New Active Cases'])
plt.plot(df['Total Active Cases'])


plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Data')

plt.legend(['New Cases','Total Cases','New Death','Total Death','New Recovered','Total Recovered','New Active Cases','Total Active Cases'])
limits=[0,50,0,10]
plt.axis(limits)
plt.show()
```

```python
plt.plot(df['New Cases'])
plt.plot(df['Total Cases'])

plt.plot(df['New Deaths'])
plt.plot(df['Total Deaths'])

plt.plot(df['New Recovered'])
plt.plot(df['Total Recovered'])

plt.plot(df['New Active Cases'])
plt.plot(df['Total Active Cases'])


plt.xlabel('Number')
plt.ylabel('Sepal Length')
plt.title('Data')

plt.legend(['New Cases','Total Cases','New Death','Total Death','New Recovered','Total Recovered','New Active Cases','Total Active Cases'],loc='upper right',ncol=2)
limits=[0,50,0,10]
plt.axis(limits)
plt.show()
```
